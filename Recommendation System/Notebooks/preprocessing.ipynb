# Preprocessing the data

```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("MovieGenre.csv")
data.columns = map(str.lower, data.columns)
data.rename(columns={'genre':'genres','imdbid':'movieId'},inplace=True)
data.dropna(inplace=True)
data.columns = data.columns.str.replace(' ', '_')
data.set_index('movieId',inplace=True)
print(data.shape)
data.drop_duplicates(inplace=True)
print(data.shape)
data.head()
```

## Preparing Genres.csv

```python
data3=data.copy()
data3.drop(['imdb_link','title','imdb_score','poster'],1,inplace=True)
data3['genres']=data3.genres.str.split('|')
```

```python
genres_df=data3.copy()
print(genres_df.head())
for index,row in data.iterrows():
    for genre in row['genres']:
        genres_df.at[index,genre]=1
genres_df.drop('genres',1,inplace=True)
genres_df.fillna(0,inplace=True)
genres_df.to_csv('genres.csv',index=True)
genres_df=pd.read_csv('genres.csv')
genres_df.head()
```
